```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = cv2.imread('photo.jpg')
```

```python
plt.imshow(img[:,:,::-1])
```

```python
np.max(img)//2
```

```python
canny = cv2.Canny(img, 127,127)
```

```python
plt.imshow(canny, cmap = 'gray')
```

```python
blur = cv2.blur(img,(3,3))
```

```python
plt.imshow(blur[:,:,::-1])
```

```python
#pass median value
median = np.median(img)
low = int(max(0,0.7*median))
high = int(max(255,1/3*median))
```

```python
print(low, median , high)
```

```python
canny_blur = cv2.Canny(blur,low,high)
```

```python
plt.imshow(canny_blur, cmap='gray')
```
